```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

plt.style.use("default")
```

```python
Results_v15 = pd.DataFrame(
    data = [
        [0.05, 0.20613789601969512, 0.02832],
        [0.1, 0.2409757037313054, 0.03043],
        [0.15, 0.23985875697593018, 0.03565],
        [0.2, 0.21970624945838146, 0.02624],
        [0.25, 0.23083915299999536, 0.02695],
        [0.3, 0.22647751932816307, 0.02575]],
    columns = ["Window size", "Total R squared", "Predictive R squared"]
)
```

```python
fig, (ax1, ax2) = plt.subplots(2)
sns.lineplot(data = Results_v15, x = "Window size", y = "Total R squared", ax = ax1, color = [0, 22/265, 189/265])
ax1.set_xlabel("")
ax1.yaxis.set_major_formatter(PercentFormatter(1, decimals = 0))
ax1.set_xticks([])
ax1.set_ylabel("$R^2_{total}$", rotation = 0, labelpad = 20)

sns.lineplot(data = Results_v15, x = "Window size", y = "Predictive R squared", ax = ax2, color = [0, 22/265, 189/265])
ax2.set_xlabel("Training window size")
ax2.yaxis.set_major_formatter(PercentFormatter(1, decimals = 1))
ax2.set_xticks([0.05, 0.1, 0.15, 0.2, 0.25, 0.3])
ax2.set_xticklabels(["5%", "10%", "15%", "20%", "25%", "30%"])
ax2.set_ylabel("$R^2_{pred}$", rotation = 0, labelpad = 20)
plt.tight_layout()
plt.savefig("../Plots/Impact window size.png", transparent = True, dpi = 600)
```
